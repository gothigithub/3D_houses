## Importing the libraries

```python
import os
import numpy as np
import pandas as pd
import rasterio as rt
import matplotlib.pyplot as plt
%matplotlib inline
```

## DEFINE DSM & DTM located in "DATA" folder

```python
DSM ='DATA\DSM_01.tif'
DTM ='DATA\DTM_01.tif'
```

# DSM

```python
dsm = rt.open(DSM)
dsm_array = dsm.read(1)
dsm_array
```

```python
plt.imshow(dsm_array)
```

# DTM

```python
dtm = rt.open(DTM)
dtm_array = dtm.read(1)
dtm_array
```

```python
plt.imshow(dtm_array)
```

# CANOPE HEIGHT MODEL

```python
#Canope Height Model = DSM - DTM
chm = dsm_array - dtm_array 
chm
```

```python
plt.imshow(chm)
plt.show()
```

```python
plt.imshow(chm,cmap='Greens')
plt.show()
```
